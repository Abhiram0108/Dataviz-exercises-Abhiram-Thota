```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: /content/gapminder.csv

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
# Task 1
europe_df = df[(df['continent'] == 'Europe') & (df['year'].isin([2002, 2007]))].copy()

fig = px.scatter(
    europe_df,
    x="gdpPercap",
    y="lifeExp",
    color=europe_df['year'].astype(str),
    hover_name="country",
    log_x=True,  # <--- FIX: Rule 4 applied
    title="Consistent Gains in Health and Wealth Across Europe (2002–2007)", # <--- FIX: Rule 5 applied
    labels={"gdpPercap": "GDP per Capita ($)", "lifeExp": "Life Expectancy (Years)", "color": "Year"},
    template="plotly_white"
)

fig.update_traces(marker=dict(size=10, opacity=0.8))
fig.show()
```

Between 2000 and 2007, most Asian countries showed clear progress in both economic and health indicators.
This is reflected in rising GDP per capita alongside increasing life expectancy.
However, the rate of improvement differed across countries, indicating uneven growth.
Some nations advanced rapidly, while others experienced slower development.
Overall, the trend highlights regional progress with noticeable disparities.

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
# Task 2
import pandas as pd
import plotly.express as px

# Load data
df = pd.read_csv('gapminder.csv')

# Choose continent to highlight
highlight_continent = 'Asia'

# Create color column
df['Highlight'] = df['Continent'].apply(
    lambda x: highlight_continent if x == highlight_continent else 'Other'
)

# Custom colors: highlight + grey
color_map = {
    highlight_continent: '#1f77b4',
    'Other': '#DDDDDD'
}

# Bubble chart
fig = px.scatter(
    df,
    x='GDP_per_capita',
    y='Life_expectancy',
    size='Population',
    color='Highlight',
    color_discrete_map=color_map,
    hover_name='Country',
    log_x=True,
    title=f'{highlight_continent} shows wide economic diversity but generally strong life expectancy outcomes',
    labels={
        'GDP_per_capita': 'GDP per Capita (log scale)',
        'Life_expectancy': 'Life Expectancy'
    }
)

# Add annotation
fig.add_annotation(
    x=15000,
    y=78,
    text='Asian countries span a wide GDP range<br>while maintaining relatively high life expectancy',
    showarrow=True,
    arrowhead=2
)

fig.update_layout(
    template='plotly_white',
    showlegend=False
)

fig.show()
```

Asia holds a large share of the world’s population, yet countries differ significantly in income levels and health conditions. Europe, on the other hand, is largely concentrated in the high-income and high life expectancy range, reflecting stable and uniform development. South American nations tend to fall in the mid-income category but still achieve comparatively strong life expectancy outcomes. This suggests that some regions are able to convert resources into better health more efficiently. Overall, global patterns reveal both regional strengths and noticeable disparities.
